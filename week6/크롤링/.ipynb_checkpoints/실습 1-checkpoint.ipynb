```python
import requests # requests 모듈 불러오기
URL = 'https://movie.naver.com/movie/bi/mi/point.nhn?code=167697' # 네이버영화에서 '신과함께'를 불러오기

# URL HTTP 요청
response = requests.get(URL) 

# HTTP Status 가져오기 (200: 정상, 404 면 File Not Found )
status = response.status_code

# HTML 소스 가져오기  
html = response.text   ###여기까진 동일!!
```

```python
from bs4 import BeautifulSoup 
movielistURL = 'https://movie.naver.com/movie/running/current.nhn' # 네이버영화에서 영화목록을 불러오기 
response = requests.get(movielistURL) 
html = response.text 

soup = BeautifulSoup(html, 'html.parser') # html.parser를 사용해서 soup에 넣기
soup.select('dt[class=tit]') #dt[class=tit]인 아이들을 선택
```

```python
for tag in soup.select('dt[class=tit]'):
    print(tag.text.strip()) # 공백을 제거하는 .strip()을 사용
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
# 태그 숫자를 제한하려면? 

## 하나만 검색 : find(‘태그명’)
html_ul = soup.find('ul')
print(html_ul.prettify())

## 모두 검색 : find_all(‘태그명’)
html_meta = soup.find_all('li')
for i in html_meta:
    print(i)
    
## 숫자 제한 해서 검색 : find_all(‘태그명’, limit=숫자)
html_meta = soup.find_all('meta', limit=3)
for i in html_meta:
    print(i)
```

```python
# 그밖에 검색 방법

## 클래스(class) 로 검색
html_class = soup.find_all(class_='tit')
for i in html_class:
    print(i)
    
## 아이디(id) 로 검색
html_id = soup.find_all(id = 'content')
for i in html_id:
    print(i)
    
## 속성으로 검색
### class라는 속성이 값이 "tit"인 녀석들을!!
html_attr = soup.find_all(attrs = {'class':'tit'})
for i in html_attr:
    print(i)

### id라는 속성의 값이 "content"인 녀석들을!!
html_attr = soup.find_all(attrs = {'id':'content'})
for i in html_attr:
    print(i)

### type라는 속성의 값이 있는 녀석들을!!
html_attr = soup.find_all(attrs={'type': True})
for i in html_attr:
    print(i)
    
## 태그의 텍스트(내용) 만 검색
html_ul = soup.find('ul')
print(html_body_ul.text)
```
