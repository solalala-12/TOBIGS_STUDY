```python
from __future__ import division
import random
import requests
import datetime as dt
import json
from functools import partial
from multiprocessing.pool import Pool

from twitterscraper.tweet import Tweet
#from twitterscraper.logging import logger
```

```python
#  query_tweets(query, limit = None, begindate = dt.date(2006,3,21), enddate=dt.date.today(), poolsize=20, lang='')
if __name__ == '__main__':
    list_of_tweets = query_tweets("장기하", begindate = dt.date(2019,2,10) , poolsize =50)

    #print the retrieved tweets to the screen:
    for tweet in query_tweets("장기하", begindate = dt.date(2019,2,10), poolsize = 50):
        print(tweet)
```

```python
text = []
time = []
```

```python
for i in range(len(list_of_tweets)) :
    text += [{"text" : list_of_tweets[i].text}]
    time += [{"time" : list_of_tweets[i].timestamp}]
```

```python
text_df = pd.DataFrame(text)
time_df = pd.DataFrame(time)

tweet_list = pd.concat([text_df,time_df],axis=1)
```

```python
tweet_list.sort_values('time')
```

```python
tweet_list.to_csv("tweet_mangone.csv")
```
