```python
# .\#title
import requests 
import re
import pandas as pd
# 단순 parsing으로 나오지 않는 Total Audience, 셀레늄 이용.
import os   
os.getcwd()
from selenium.webdriver import Chrome

from tqdm import tqdm
```

```python

url="https://movie.daum.net/boxoffice/yearly?year=2018"
res = requests.get(url, headers = {
    'user-agent': ('Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) '
                   'Chrome/41.0.2228.0 Safari/537.36')})
res
```

```python
html=res.text
# url 에서 얻어온 정보 다 가져오기
html
```

```python
# Beautiful Soup를 이용하여, html 언어로 가져오기
from bs4 import BeautifulSoup 
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
# a태그의, class name이 name_movie  50개만 추출!
names=soup.find_all('a',{'class':'name_movie #title'},limit=50)
```

```python
# 이 영화의 제목을 name_list에 넣어준다.
name_list=[]
for i in names:
    name_list.append(i.get_text().strip())
```

```python
#len(name_list)
name_list[:10]
```

```python
# 개봉일자, 평점, 누적관객수, 장르 
# 각 영화에 대한 url을 가져와서 그 url로 접근하여, 4가지 요소를 가져온다.
link_list=[]
linkes=soup.find_all('a',{'class':'name_movie #title'},limit=50)
for i in linkes:
    link_list.append(i.get('href'))
```

```python
#범블비 주소 오류나서 수동 지정
link_list[15]='/moviedb/main?movieId=102301'
```

```python
# 장르, 개봉날짜, 평점 추출 

genre_list=[]
date_list=[]
rating_list=[]

for i in tqdm(link_list):
    url='https://movie.daum.net'+i
    #print(url)
    res = requests.get(url, headers = {
    'user-agent': ('Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) '
                   'Chrome/41.0.2228.0 Safari/537.36')})
    html=res.text
    soup=BeautifulSoup(html,'html.parser')
    genre_list.append(soup.find_all('dd',{'class':'txt_main'})[0].text)
    date_list.append(soup.find_all('dd',{'class':'txt_main'})[1].text.strip()[:10])
    rating_list.append(soup.find('em',{'class':"emph_grade"}).text)
    
    
    
print("Finish!")
    #print('genre: %s date: %s rating: %s ta %s'%(genre,date,rating,ta))
```

```python
# 누적 관개수는 셀레니움 이용
ta_list=[]
browser=Chrome()
for i in tqdm(link_list):
    url='https://movie.daum.net'+i
    
    browser.get(url)
    html=browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    notices = soup.select('#totalAudience')[0].text
    ta_list.append(notices)
    browser.close()
```

```python
data={'name':name_list,'date':date_list,'rating':rating_list,'ta':ta_list,'genre':genre_list}
```

```python
df=pd.DataFrame(data)
```

# 데이터 탐색

```python
# 어차피 2018 년 자료니 월만 가져온다.
df['date']=df['date'].str.slice(5,7)
```

```python
df[:10]
```

```python
df['rating']=df['rating'].astype(float)
```

```python
# 누적 관객에서 정규식을 이용해, 숫자만 가져온다.
df['ta'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
```

```python
df['ta']=df['ta'].astype(int)
```

```python
df.info()
```

```python
# 총 몇개의 unique한 장르가 있는지 살펴본다.
# 장르가 세분화 되어있음으로 대 장르로 묶는다.
# / 에서 첫번째 단어만 추출
f = lambda x:" ".join(x for x in x.split('/')[0])
df['genre'] = df['genre'].apply(f)
```

```python
import matplotlib.pyplot as plt  
```

```python
# 평점과 누적 관관객 수 사이의 관계를 보기 위해 scatter plot을 그려보았는데
# 어느정도 상관성이 있어보인다.
plt.xlabel('rating')
plt.ylabel('Total Audience')
plt.scatter(df['rating'],df['ta'])
plt.show()
```
