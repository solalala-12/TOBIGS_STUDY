## 1. 영문 대소문자로 이루어진 문자열 입력했을 때, 문자열 내에 몇개의 단어가 있는지 구하는 알고리즘을 작성하시오.
(중복된 단어는 제외한다)

```python
import re
print("문자열을 입력하세요")

t=list()
x=input()

x_s=x.split(" ")




for i in x_s:
    #와. 등 쓸데없는 문자 제거! 온전히 문자열만
    # [ ] 안에 제거할 요소들을 넣어주면 된다. 
    a= re.sub("[(*&.,]", "",i)
    print(a)
    if(a in t):
        continue
    else:
        t.append(a)
len(t)
```

## 2) 대문자로 이루어진 문자열을 입력하였을 때 특정 열이 몇번 등장하는가?

```python
print("(문자열"+", 찾을 문자열)"+" 형식으로 입력하세요")
x1=input()

base=x1.split(',')[0]
find=x1.split(',')[1]

print(find,"가 나타난 횟수:", base.count(find))
```

## 3) 숫자와 알파벳으로 구성된 문자열을 입력받아서 숫자만큼 알파벳을 반복시켜 반환하는 함수이다. 이 때 1~9 사이의 숫자 뒤에 알파벳이 하나 나오는 패턴이반복된다. 그 기능을 수행하는 함수 uncompress(s)를 구현하시오

```python
# 문자 따로 숫자 따로 저장한다.
#저장한 다음에 각각의 인덱스로 접근해서 출력
print("문자열을 입력하세요")
n=list()
c=list()
x3=input()
for i in x3:
    if(i.isdigit()):
        n.append(i)
    else:
        c.append(i)
    
    

f=""
for i in range(0,len(n)):
    f=f+(c[i]*int(n[i]))
    
f
```

## 5) 문자열을 입력하면 입력된문자열과 anagram 관계인 문자열을 출력하고 총 몇개인지도 출력하라.

```python
import itertools
print("문자열을 입력하세요")
x5=input()
```

```python
x6=list(x5)
x6
#가능한 모든 순열을 만들어주는 package

list_anagram =  itertools.permutations(x6)
for i in list_anagram:
    print(i)
```
