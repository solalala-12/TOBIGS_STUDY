## 1-1)날짜 변수 처리 
-경매 분석상 날짜 데이터가 많으니 오늘날짜 기준 며칠이 지났는지 계산하는 함수 정의

```python

def change_datetime(x):
    today = datetime.date.today()
    someday = datetime.date(int(x[:4]), int(x[5:7]), int(x[8:10]))
    diff = today - someday
    return diff.days
```

```python
train["Appraisal_date"] = train["Appraisal_date"].apply(lambda x : x[:10])
train["First_auction_date"] =train["First_auction_date"].apply(lambda x : x[:10])
train["Final_auction_date"] =train["Final_auction_date"].apply(lambda x : x[:10])
train["Preserve_regist_date"] =train["Preserve_regist_date"].apply(lambda x : x[:10])
train["Close_date"] = train["Close_date"].apply(lambda x : x[:10])
train["Appraisal_date"] = train["Appraisal_date"].apply(change_datetime)
train["First_auction_date"] = train["First_auction_date"].apply(change_datetime)
train["Final_auction_date"] =train["Final_auction_date"].apply(change_datetime)
train["Preserve_regist_date"] = train["Preserve_regist_date"].apply(change_datetime)
train["Close_date"] = train["Close_date"].apply(change_datetime)
```

## svm

```python
y=np.array(y)
x=np.array(x)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

def fivefold(x,y,model):
    model_scores = np.zeros(5)
    scores = np.zeros(5)

    cv = KFold(5, shuffle=True, random_state=0)
    for i, (idx_train, idx_test) in enumerate(cv.split(x)):
        #데이터 하나씩 트레인과 테스트에 부과
        x_train = x[idx_train]
        x_test = x[idx_test]
        y_train = y[idx_train]
        y_test = y[idx_test]

        #모델 fit
        result = model.fit(x_train,y_train)

        #prediction
        model_pred = result.predict(x_train)
        pred = result.predict(x_test)
        model_accuracy = accuracy_score(y_train, model_pred)
        accuracy= accuracy_score(y_test, pred)

        #test 5번 하나하나 프린트
        model_scores[i] = model_accuracy
        scores[i] = accuracy
        print("학습 accuacy = {:.8f}, 검증 accuacy = {:.8f}".format(model_accuracy,accuracy))
    print('평균 train accuacy : ' + str(model_scores.mean()) )
    print('평균 test accuacy : ' + str(scores.mean()) )
```

```python
x
```

```python
rbf_kernel_svm_clf = SVC(kernel="rbf",C=0.001,gamma=0.5)

rbf_kernel_svm_clf.fit(x, y)
```

```python
# 최적의 C 와 gamma 를 찾기 위해 grid search 수행
from sklearn.model_selection import GridSearchCV
values={'C':[0.001,0.1,1,10],'gamma':[0.5,5,10,50]}
```

```python
grid_search=GridSearchCV(estimator=rbf_kernel_svm_clf,param_grid=values,cv=5)
```

```python
grid_search
```

```python

# 그리디 서치 적용 (최적의 파라미터 찾기 위함)
grid_search.fit(x,y)
```

```python
grid_search.best_score_
```

```python
# 최적의 paramator 는  c=10 과 gamma 50.
grid_search.best_params_
```

```python
rbf_kernel_svm_clf = SVC(kernel="rbf",C=10,gamma=50)

rbf_kernel_svm_clf.fit(x, y)
```

```python
# 5 교차 검증을 한 정확도

fivefold(x,y,rbf_kernel_svm_clf)
```

```python
# 정확도가 0.95정도로 높은 편이다. 
```
